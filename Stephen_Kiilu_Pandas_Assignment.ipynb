**Import python packages and data to be used**

```python
import pandas as pd
import numpy as np
```

```python
west=pd.read_csv('/content/af_west.csv',skipinitialspace = True)
north=pd.read_csv('/content/af_north.csv',skipinitialspace = True)
central=pd.read_csv('/content/af_central.csv',skipinitialspace = True)
east=pd.read_csv('/content/af_east.csv',skipinitialspace = True)
south=pd.read_csv('/content/af_austral.csv',skipinitialspace = True)
```

**Part 1**

**Q1: Print the first 5 observations**

```python
west.head(5)
east.head(5)
south.head(5)
north.head(5)
central.head(5)
```

```python
dat=[west,east,south,north,central]
data=pd.concat(dat,axis=0,ignore_index=True)
data['Region']='w'
data
```

**Q2 Merging data files**





```python
data.loc[0:15,'Region']='West'
data.loc[16:27,'Region']='East'
data.loc[28:39,'Region']='South'
data.loc[40:44,'Region']='North'
data.loc[45:53,'Region']='Central'
```

```python
data
```

**Q3: Calculate mean, std deviation and coefficient of variation**

```python
data.groupby('Region').agg(
    mean_population=('Population', 'mean'),
    std_population=('Population', 'std'),
    coefficient_variation=("Population", lambda x: (x.std()/x.mean())))


```

**Part 2**

**Q1: Minimum area of each region and the corresponding country**

```python
data.groupby('Region').apply(lambda p:p.loc[p.Area.idxmin()]).loc[:,['Names','Area']]
```

**Q2: Maximum area of each region and the corresponding country**

```python
data.groupby('Region').apply(lambda p:p.loc[p.Area.idxmax()]).loc[:,['Names','Area']]
```

**Q3: Minimum population of each region and the corresponding country**

```python
data.groupby('Region').apply(lambda p:p.loc[p.Population.idxmin()]).loc[:,['Names','Population']]
```

**Q4: Maximum population of each region and the corresponding country**

```python
data.groupby('Region').apply(lambda p:p.loc[p.Population.idxmax()]).loc[:,['Names','Population']]
```

**Q5: Average population and country with population close to mean**

```python
data.groupby('Region').Population.agg(['mean'])
```

```python
data.groupby(["Region"]).apply(lambda y:y.loc[(y.Population-y.Population.mean()).abs().idxmin()]).loc[:,['Names','Population']]
```

```python
 regions=data.groupby('Region')
Popopulation_average = data.loc[((regions['Population'].transform('mean') - data['Population']).abs().groupby(data['Region']).idxmin()), ['Names','Population','Region']]
Popopulation_average
```

**Q6: Average population and country with population furthest from mean**

```python
data.groupby(["Region"]).apply(lambda p:p.loc[(p.Population-p.Population.mean()).abs().idxmax()]).loc[:,['Names','Population']]
```

**Q7: Average density per region and region closest to the average density.**

```python
data.groupby(["Region"]).Density.agg('mean')
data.groupby(["Region"]).apply(lambda p:p.loc[abs(p.Density-p.Density.mean()).idxmin()]).loc[:,['Names','Density']]
```

**Q8: Average density per region and region furthest from the average density.**

```python
data.groupby(["Region"]).Density.agg('mean')
data.groupby(["Region"]).apply(lambda p:p.loc[abs(p.Density-p.Density.mean()).idxmax()]).loc[:,['Names','Density']]
```
